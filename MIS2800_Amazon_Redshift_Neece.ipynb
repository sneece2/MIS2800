2. Speed to Insight: How does the architecture of a "Leader Node" and "Compute Nodes" change how quickly a Business Intelligence report is generated?  https://docs.aws.amazon.com/redshift/latest/dg/c_high_level_system_architecture.html


## YOUR ANSWERS TO PART 2 BELOW

Question 1: Redshift’s columnar storage stores data by column rather than by row, meaning analytics queries can read only the specific columns they need and reduce disk I/O with better compression; this is far more impactful for large‑scale analytical workloads than perfect normalization, which is more about transactional integrity than fast aggregated reads.


Question 2: With a leader node that parses queries and plans work, and many compute nodes that execute pieces of those plans in parallel (MPP) before aggregating results, Redshift can distribute heavy analytical workloads across CPUs and disks, significantly speeding up BI report generation
