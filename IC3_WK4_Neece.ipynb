```python
#Block 1
# Instructions: Create a table named stock_performance with columns for ticker, price_before_game, price_week_after_
%%sql
CREATE TABLE IF NOT EXISTS stock_performance (
    ticker TEXT,
    price_before_game REAL,
    price_week_after_game REAL,
    trading_volume_millions INTEGER
) STRICT;
```

```python
# Instructions: Create a table named Ad_Buys with columns for ticker, company_name, ad_cost_millions, airtime_second
# The ticker should be a TEXT data type and serve as the Primary Key.
# Select an appropriate data_type for the other columns.
# Remember to use the STRICT keyword to ensure the datatype is enforced.
%%sql
CREATE TABLE IF NOT EXISTS Ad_Buys (
    ticker TEXT PRIMARY KEY,
    company_name TEXT,
    ad_cost_millions REAL,
 airtime_seconds INTEGER
) STRICT;
```
