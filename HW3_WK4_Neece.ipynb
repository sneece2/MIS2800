```python
#Block 1
# Instructions: There is a typo in the table.  Hint: a word is missing.
%%sql
CREATE TABLE Marketing_Brands (
  brand_id INTEGER PRIMARY KEY,
  brand_name TEXT,
  brand_type TEXT,
  brand_budget INTEGER
) STRICT;
```

```python
#Block 3
# -- Instructions: Fix the code to select all the columns
# from the Marketing_Brands table
%%sql
SELECT * FROM Marketing_Brands;
```

```python
#Block 4
# Instructions: Fix the code to display only the brand names and brand types that had budget over 700000
%%sql
SELECT brand_name, brand_type FROM Marketing_Brands;
```

```python
#Block 5
%%sql
-- Instructions: Fix the code to find the brand types that are Religious
SELECT * FROM Marketing_Brands WHERE brand_type = 'Religious'
```

```python
#Block 6
%%sql
-- Instructions: Fix the code to display the brand names that have a budget under 1000000
SELECT * FROM Marketing_Brands WHERE brand_budget < 1000000
```

```python
#Block 7
# Instructins: Fix the code so that this will run SQLite3 in
# Colab without having to wrap in Python.  Hint: It's Magic!
%%sql
SELECT * FROM Marketing_Brands WHERE brand_budget > 7000000
```

```python
#Block 8
%%sql
--Instructions: Find the brands where the brand type is Snacks or Beverages.  Display only the brand names.
SELECT brand_name FROM Marketing_Brands where brand_type = 'Snacks' or 'Beverages'
```

```python
#Block 9
%%sql
-- Instructions: Sort the table by brand budget in DESC order. Display only the brand_name and brand_budget.
SELECT brand_name, brand_budget FROM Marketing_Brands ORDER BY brand_budget DESC
```
