```python
%%sql
CREATE TABLE products (
    product_id INTEGER PRIMARY KEY,
    name TEXT,
    price REAL
) STRICT;
```

```python
%%sql
INSERT INTO products (name, price) VALUES ('Superbowl Jersey', 120.00);
INSERT INTO products (name, price) VALUES ('Football Cleats', 85.50);
INSERT INTO products (name, price) VALUES ('Team Hat', 35.00);
```

```python
# Show all products that cost more than $50
%%sql
SELECT * FROM products WHERE price > 50;
```

```python
# Update a Price
%%sql
UPDATE products
SET price = 99.99
WHERE name = 'Superbowl Jersey';
```

```python
# Delete an Item
%%sql
DELETE FROM products
WHERE name = 'Team Hat';
```
